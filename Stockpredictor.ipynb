```python
#Split the data into train and test sets
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size=0.5)
```

```python
#model
def gradient_descent(m1_now, m2_now, m3_now, b_now, points, L):
  m1_gradient = 0
  m2_gradient = 0
  m3_gradient = 0
  b_gradient = 0

  n = len(points)

  for i in range(n):
    x1 = points.iloc[i].open
    x2= points.iloc[i].high
    x3= points.iloc[i].low
    y = points.iloc[i].close


    m1_gradient += -(2/n) * x1 * (y - (m1_now * x1 + m2_now * x2 + m3_now * x3 + b_now))
    m2_gradient += -(2/n) * x2 * (y - (m1_now * x1 + m2_now * x2 + m3_now * x3 + b_now))
    m3_gradient += -(2/n) * x3 * (y - (m1_now * x1 + m2_now * x2 + m3_now * x3 + b_now))
    b_gradient += -(2/n) * (y - (m1_now * x1 + m2_now * x2 + b_now))

  m1 = m1_now - m1_gradient * L
  m2 = m2_now - m2_gradient * L
  m3 = m3_now - m3_gradient * L
  b = b_now - b_gradient * L
  return m1, m2, m3, b
```

```python
#Train the model with train data

m1 = 0
m2 = 0
m3 = 0
b = 0
L = 0.000001
epochs = 30

for i in range(epochs):
  m1, m2, m3, b = gradient_descent(m1, m2, m3, b, train, L)
print(m1, m2, m3, b)
```

```python
def calculate_result(m1, m2, m3, b, points):
  n = len(points)

  array = []
  for i in range(n):
    x = points.iloc[i].open
    x2 = points.iloc[i].high
    x3 = points.iloc[i].low
    array.append(m1 * x + m2 * x2 + m3 * x3 + b)
  return array
```

```python
result = calculate_result(m1, m2, m3, b, test)
```

```python
test['Predictions'] = result
test.head(50)
```

```python
def percentage_change(col1,col2):
    return ((col2 - col1) / col1) * 100

test['error'] = -1 * percentage_change(test['close'],test['Predictions'])
average_error = test['error'].mean()
average_error
```

```python
result_old = calculate_result(m1, m2, m3, b, train)
train['Predictions'] = result_old
train['error'] = percentage_change(train['close'],train['Predictions'])
average_error_old = train['error'].mean()
average_error_old
```
